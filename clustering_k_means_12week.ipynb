```python
import numpy as nm
import matplotlib.pyplot as plt
import pandas as pd
# download dataset from https://gist.github.com/pravalliyaram/5c05f43d2351249927b8a3f3cc3e5ecf
dataset = pd.read_csv("Mall_Customers.csv")

dataset.head()
```

```python
x = dataset.iloc[:, [3,4]]
x.head()
```

```python
# importing the k means algorithm
from sklearn.cluster import KMeans
```

```python
WCSS_list = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init= "k-means++",random_state= 42)
  kmeans.fit(x)
  WCSS_list.append(kmeans.inertia_)
# print(WCSS_list)

# lets draw the graph
plt.plot(range(1, 11), WCSS_list)
plt.xlabel("Kume sayisi")
plt.ylabel("WSS score ")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Initialize the KMeans model with 5 clusters
# init="k-means++": Initialization method to place the initial cluster centroids efficiently
# random_state=42: Fixing the random state ensures reproducible results
kmeans = KMeans(n_clusters=5, init="k-means++", random_state=42)

# Fit the KMeans model to the dataset 'x' and predict the cluster for each sample
y_predict = kmeans.fit_predict(x)

# Print the cluster assignments for each sample in 'x'
print(y_predict)
```

```python
plt.scatter(x.iloc[y_predict == 0, 0], x.iloc[y_predict == 0, 1], s = 100, c='blue', label = "cluster_1")
plt.scatter(x.iloc[y_predict == 1, 0], x.iloc[y_predict == 1, 1], s = 100, c='red', label = "cluster_2")
plt.scatter(x.iloc[y_predict == 2, 0], x.iloc[y_predict == 2, 1], s = 100, c='green', label = "cluster_3")
plt.scatter(x.iloc[y_predict == 3, 0], x.iloc[y_predict == 3, 1], s = 100, c='yellow', label = "cluster_4")
plt.scatter(x.iloc[y_predict == 4, 0], x.iloc[y_predict == 4, 1], s = 100, c='purple', label = "cluster_5")

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s = 300, c ='pink', label = "Centroid")

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
