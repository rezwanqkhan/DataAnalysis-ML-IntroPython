```python
import numpy as np
import matplotlib.pyplot as mtp
import pandas as pd
```

```python
# you can dowload the following dataset here https://github.com/aocdiken/Dataset/blob/main/Mall_customer.csv
dataset = pd.read_csv("Mall_customer.csv")
# first 5
dataset.head()
```

```python
# take the 3rd ans 4rd column with all rows to the varible x
x = dataset.iloc[:,[3,4]].values
print(x)
```

```python
# imprting the hierarchy
import scipy.cluster.hierarchy as shc
# by using the dendrogram we determine that in how many cluster it shoul be
dendro = shc.dendrogram(shc.linkage(x, method="ward"))
mtp.xlabel("customers")
mtp.ylabel("Distances")
mtp.show()
```

```python
# by using the dendrogram we determined that the n_clusters could be 5
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_pred = hc.fit_predict(x)
print(y_pred)
```

```python
mtp.scatter(x[:,0], x[:,1], s = 50, c = 'black')
mtp.xlabel("Yllk Gelir")
mtp.ylabel("Harcama Skoru")
mtp.show()
```

```python
# lets use the y_predict in diffrent color for different clusters
mtp.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], s = 50, c ='blue', label = 'Kume 1')
mtp.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], s = 50, c ='red', label = 'Kume 2')
mtp.scatter(x[y_pred == 2, 0], x[y_pred == 2, 1], s = 50, c ='green', label = 'Kume 3')
mtp.scatter(x[y_pred == 3, 0], x[y_pred == 3, 1], s = 50, c ='orange', label = 'Kume 4')
mtp.scatter(x[y_pred == 4, 0], x[y_pred == 4, 1], s = 50, c ='pink', label = 'Kume 5')
mtp.xlabel("Yllk Gelir")
mtp.ylabel("Harcama Skoru")
mtp.show()
```
